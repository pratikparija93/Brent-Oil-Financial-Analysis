#### Defining the OLS function

```python
def ols(Y, X):
    beta_hat = dot(dot(inv(dot(X.T, X)), X.T), Y)#inv(X.T * X) * X.T * Y
    y_hat = np.matmul(X, beta_hat) #X * beta_hat
    residuals = Y - y_hat
    print("residuals:")
    print(residuals)
    
    from numpy import sqrt
    sigma2 = (1 / 251) * np.dot(residuals.T,residuals)    #residuals.T * residuals
    sigma = sqrt(sigma2)
    print("sigma:")
    print(sigma)
    
    varcov_beta_hat2 = inv(dot(X.T,X))     #(sigma^2)*inv(X.T *X)
    print("varcov_beta_hat2")
    print(varcov_beta_hat2)#varcov_beta_hat = (sigma^2)*varcov_beta_hat2

    varcov_beta_hat = sigma*sigma*varcov_beta_hat2
    print("varcov_beta_hat")
    print(varcov_beta_hat)

    var_beta_hat = np.diag(varcov_beta_hat)
    print("var_beta_hat")
    print(var_beta_hat)
    
    a = np.dot(residuals.T,residuals)
    b = np.var(Y)
    print(a)
    print(b)
    R_square = 1 - ((np.dot(residuals.T,residuals)) / (np.var(Y)*251))
    print("R_square")
    print(R_square)
    
    adjust_R_square = 1-((1-R_square)*(251-1)/(251-4-1))
    print("adjust_R_square")
    print(adjust_R_square)

    print("beta_hat")
    print(beta_hat)
    
    print("var_beta_hat")
    print(var_beta_hat)
    
    print("np.std(var_beta_hat)")
    print(np.std(var_beta_hat))
    
    t_stat = beta_hat / (np.std(var_beta_hat))
    print("t_stat")
    print(t_stat)
    
    from scipy.stats import norm
    p_val_t = 1-norm.cdf(t_stat)
    print("p_val_t")
    print(p_val_t)
    
    F_stat = np.dot(dot(beta_hat.T, varcov_beta_hat),beta_hat)/(np.dot(residuals.T,residuals))
    print("F_stat")
    print(F_stat)
    
    from scipy.stats import chi2
    p_val_F = 1-chi2.cdf(F_stat,250)
    print("p_val_F")
    print(p_val_F)
    
```

```python
ols(Y, X)
```
