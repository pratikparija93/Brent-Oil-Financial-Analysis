```python
S = len(dataset.index)
result_gasmomentum = dict()
for i in range(0, S, 20):
    if i + 20 >= S:
        break
    y = dataset.iloc[i + 20]['DHHNGSP'] - dataset.iloc[i]['DHHNGSP']
    date1 = dataset.iloc[i + 20]['observation_date']
    result_gasmomentum[date1] = y
    arr1=list(result_gas.items())
    momentum_gas = df(list(result_gasmomentum.items()),columns=['Date','Momentum of gas price'])
```

```python
def change(y):
    
    if y<0:
        return -1
    if y>0:
        return 1
```

```python
momentum_gas["Trading Signal"]=""
```
